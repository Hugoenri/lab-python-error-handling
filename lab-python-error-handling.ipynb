def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")

````python
def calculate_total_price(customer_orders):
    total_price = 0.0
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}:"))
                if price < 0:
                    raise ValueError("The price cannot be negative. Please enter a valid price.")
                total_price += price
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
    return total_price
````

````python
def get_customer_orders(inventory):
    while True:
        try:
            num_orders = int(input("Enter the number of customer orders: "))
            if num_order < 0:
                raise ValueError("The number of orders cannot be negative.")
            break  
        except ValueError as error:
            print(f"Error: {error}. Please enter a valid number")

    orders = []
    for _ in range(num_order):
        while True:
            product = input('Enter the name of the product you want to order:')
            if product not in inventory:
                print("Error: The product is not in inventory. Try again.")
            elif inventory[product] <= 0:
                print("Error: There is no stock available for this product. Try again.")
            else:
                orders.append(product)
                break  

    return orders
````

````python
def initialize_inventory():
    inventory = {}
    while True:
        try:
            num_items = int(input("Enter the number of items in inventory: "))
            if num_items < 0:
                raise ValueError("The number of items cannot be negative.")
            break
        except ValueError as e:
            print(f"Error: {e}. Please enter a valid number.")

    for _ in range(num_items):
        product_name = input("Enter the product name: ")
        
        while True:
            try:
                quantity = int(input(f"Enter the quantity for {product_name}: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative.")
                break
            except ValueError as e:
                print(f"Error: {e}. Please enter a valid number.")
        
        inventory[product_name] = quantity

    return inventory

# Example execution
inventory = initialize_inventory()
print(inventory)
````
